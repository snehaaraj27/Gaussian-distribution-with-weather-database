```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("/content/Kolkata_weather_data(2017-2022).csv")
df
```

```python
Numerics=LabelEncoder()
```

```python
inputs=df.drop('Info',axis='columns')
target=df['Info']
target
```

```python
inputs['Dew Point_n']=Numerics.fit_transform(inputs['Dew Point'])
inputs['Temperature_n']=Numerics.fit_transform(inputs['Temperature'])
inputs['Wind Speed_n']=Numerics.fit_transform(inputs['Wind Speed'])
inputs['Cloud Cover_n']=Numerics.fit_transform(inputs['Cloud Cover'])
inputs
```

```python
inputs_n=inputs.drop(['Dew Point','Temperature','Wind Speed','Cloud Cover'],axis='columns')
inputs_nd = inputs_n.dropna(how='all')
inputs_nd
```

```python
import numpy as np

# Assuming 'target' is a NumPy array
nan_indices = np.isnan(target)

# Print indices of NaN values
print("Indices of NaN values in 'target':")
print(np.where(nan_indices))

# Print the actual NaN values in 'target'
print("NaN values in 'target':")
print(target[nan_indices])
```

```python
#
nan_values = inputs_nd[inputs_nd.isna().any(axis=1)]
print(nan_values)
```

```python
import pandas as pd
from scipy.stats import norm
column_name = 'Temperature_n'
data = inputs_nd[column_name]
mean = data.mean()
std = data.std()
pdf_values = norm.pdf(data, mean, std)
inputs_nd['pdf_values'] = pdf_values
print(inputs_nd)

```

```python
import matplotlib.pyplot as plt

mean = 0
std_values = [1, 2, 3]
x = np.linspace(-10, 10, 1000)
for std in std_values:
    y = norm.pdf(x, mean, std)
    plt.plot(x, y, label=f'Std Dev = {std}')
plt.title('Gaussian Distribution with Varying Standard Deviation')
plt.xlabel('X-axis')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```

```python
inputs_nd.plot(kind='bar', x='Temperature_n', y='Relative Humidity', legend=False)
plt.title('Bar Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()

inputs_nd.plot(kind='scatter', x='Temperature_n', y='Relative Humidity')
plt.title('Scatter Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```
